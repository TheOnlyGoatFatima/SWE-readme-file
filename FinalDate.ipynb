```python
import pandas as pd
import matplotlib.pyplot as plt

# Data preparation
data = {
    'Region': ['North America'] * 49 + ['Europe'] * 31 + ['Asia'] * 31 + ['Australia'] * 15 + ['Africa'] * 15 +
             ['North America'] * 24 + ['Europe'] * 25 + ['North America'] * 23 + ['Europe'] * 26 +
             ['Asia'] * 3 + ['Australia'] * 2,
    'Calendar': ['Semester'] * 49 + ['Semester'] * 31 + ['Semester'] * 31 + ['Semester'] * 15 + ['Semester'] * 15 +
                ['Quarter'] * 24 + ['Trimester'] * 25 + ['Other'] * 23 + ['Other'] * 26 + # Changed length from 26 to 23 to match other lists
                ['Other'] * 3 + ['Other'] * 2
}

df = pd.DataFrame(data)

# Count universities by region and calendar type
count_data = df.groupby(['Region', 'Calendar']).size().unstack().fillna(0)


# Plotting
percentage_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

ax = percentage_data.plot(kind='bar', stacked=True, figsize=(14, 8)) # Removed autopct argument
plt.title('Number of Universities by Calendar Type in Each Region')
plt.xlabel('Region')
plt.ylabel('Percentage of Universities')
plt.legend(title='Calendar Type')
plt.xticks(rotation=45)
plt.tight_layout()

# Add count labels
for i, region in enumerate(count_data.index):
    for j, calendar in enumerate(count_data.columns):
        count = count_data.loc[region, calendar]
        if count > 0:
            plt.text(i, percentage_data.iloc[i, :j].sum() + percentage_data.iloc[i, j] / 2,
                     str(int(count)), ha='center', va='center', color='white')

# Show plot
plt.show()
```
