```python
# @title Default title text
import pandas as pd
import matplotlib.pyplot as plt

# Data preparation
data = {
    'Region': ['North America'] * 49 + ['Europe'] * 31 + ['Asia'] * 31 + ['Australia'] * 15 + ['Africa'] * 15 +
             ['North America'] * 24 + ['Europe'] * 25 + ['North America'] * 23 + ['Europe'] * 26 +
             ['Asia'] * 3 + ['Australia'] * 2,
    'Calendar': ['Semester'] * 49 + ['Semester'] * 31 + ['Semester'] * 31 + ['Semester'] * 15 + ['Semester'] * 15 +
                ['Quarter'] * 24 + ['Trimester'] * 25 + ['Other'] * 23 + ['Other'] * 26 + # Changed length from 26 to 23 to match other lists
                ['Other'] * 3 + ['Other'] * 2
}

df = pd.DataFrame(data)

# Count universities by region and calendar type
count_data = df.groupby(['Region', 'Calendar']).size().unstack().fillna(0)


# Plotting
percentage_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

ax = percentage_data.plot(kind='bar', stacked=True, figsize=(14, 8)) # Removed autopct argument
plt.title('Number of Universities by Calendar Type in Each Region')
plt.xlabel('Region')
plt.ylabel('Percentage of Universities')
plt.legend(title='Calendar Type')
plt.xticks(rotation=45)
plt.tight_layout()

# Add count labels
for i, region in enumerate(count_data.index):
    for j, calendar in enumerate(count_data.columns):
        count = count_data.loc[region, calendar]
        if count > 0:
            plt.text(i, percentage_data.iloc[i, :j].sum() + percentage_data.iloc[i, j] / 2,
                     str(int(count)), ha='center', va='center', color='white')

# Show plot
plt.show()
```

## If the choice is made to proceed considering the **semester calendar** as the most comprehensive, this is a chart that shows the midterm and final weeks, which should be excluded.
### Note: The following graph does not exclude holidays, as they vary slightly between countries, and is not a definitive reason for excluding the conference at this time.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# Define the months and weeks
months = ['January', 'February', 'April']
weeks = {
    'January': [(14, 17), (21, 24)],
    'February': [(4, 7), (11, 14)],
    'April': [(7, 10), (14, 17)],
}

# Define the timeline for midterms and finals
midterms = {'February': [(25, 29)], 'March': [(1, 10), (15, 21)]}
finals = {'April': [(30, 30)], 'May': [(1, 10), (15, 31)]}

# Function to add date range patches
def add_week_patches(ax, weeks, y, color, label):
    for start_day, end_day in weeks:
        ax.add_patch(plt.Rectangle((start_day, y), end_day - start_day + 1, 0.8, color=color, label=label))

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 5))
ax.set_xlim(0, 31)
ax.set_ylim(-1, len(months))
ax.set_yticks(range(len(months)))
ax.set_yticklabels(months)
ax.set_xlabel('Days of the Month')
ax.set_ylabel('Months')

# Add grid lines
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d'))
ax.grid(True, which='both', axis='x', linestyle='--', linewidth=0.5, alpha=0.7)

# Plot suitable weeks for the conference
for i, month in enumerate(months):
    if month in weeks:
        add_week_patches(ax, weeks[month], i, color='green', label='Suitable Weeks')

# Plot midterms and finals
for month, periods in midterms.items():
    if month in months:
        add_week_patches(ax, periods, months.index(month), color='orange', label='Midterms')

for month, periods in finals.items():
    if month in months:
        add_week_patches(ax, periods, months.index(month), color='red', label='Finals')

# Add a legend
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc='upper right')

# Add title
plt.title('Suitable Weeks for Student Research Conference (2024)', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

# **5. Travel and Accommodation**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Ticket data with correct formatting for Middle East
data = {
    'Category': [
        'Local Economy', 'Local Business',
        'Middle East Economy', 'Middle East Business',
        'Global Economy', 'Global Business'
    ],
    'January Low': [250, 1000, 900, 2000, 3500, 13000],
    'January High': [550, 2500, 2500, 8500, 6500, 30000],
    'February Low': [250, 1000, 1000, 2000, 2500, 13000],
    'February High': [550, 2500, 4500, 8500, 5000, 30000],
    'April Low': [250, 1000, 850, 2000, 2000, 13000],
    'April High': [550, 2500, 2000, 8500, 4500, 30000]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Melting the DataFrame for easy plotting
df_melt = df.melt(id_vars='Category', var_name='Month', value_name='Price')

# Extracting the Month and Range type
df_melt['Month'] = df_melt['Month'].str.split(' ').str[0]
df_melt['Range'] = df_melt['Month'].str.split(' ').str[1]
df_melt['Type'] = df_melt['Category'].str.split().str[1]
df_melt['Region'] = df_melt['Category'].str.split().str[0]

# Correct the splitting and naming issues for accurate plot labels
df_melt['Month'] = df_melt['Month'].str.split().str[0]

# Function to plot price range
def plot_price_ranges(df, class_type, title, ax):
    sns.barplot(
        x='Region',
        y='Price',
        hue='Month',
        data=df[df['Type'] == class_type],
        ci=None,
        ax=ax
    )
    ax.set_title(f'{title} Class Price Comparison')
    ax.set_ylabel('Price (SAR)')
    ax.set_xlabel('Region')
    ax.set_ylim(0, 32000)
    ax.legend(title='Month', loc='upper right')

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(16, 8))
sns.set(style="whitegrid")

# Plot each class type separately
plot_price_ranges(df_melt, 'Economy', 'Economy', axs[0])
plot_price_ranges(df_melt, 'Business', 'Business', axs[1])

plt.suptitle('Ticket Price Range Comparison for 2025', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

### April emerges as the most economical month, especially for the Middle East and Global Economy categories, with some reductions in average prices.

### Local travel costs are stable, across January, February, and April.

### **Prices are very close, which makes the impact of this factor on decision-making limited. However, it increases globally in January by a small percentage.**
